# Notebook 05: Semantic Search Concepts

Use this workbook during Days 051-060. This lightweight search example uses TF-IDF so it runs locally without an API key. Later, compare it with embedding-based search.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "Docker packages an application and its dependencies into a container image.",
    "Kubernetes manages containerized workloads across a cluster.",
    "A machine learning model learns patterns from training data.",
    "A confusion matrix helps inspect classification errors.",
    "Terraform describes infrastructure using configuration files.",
]

vectorizer = TfidfVectorizer(stop_words="english")
document_vectors = vectorizer.fit_transform(documents)

def search(query, limit=3):
    query_vector = vectorizer.transform([query])
    scores = cosine_similarity(query_vector, document_vectors).flatten()
    ranked_indexes = scores.argsort()[::-1][:limit]
    return [(documents[index], round(float(scores[index]), 3)) for index in ranked_indexes]

search("How do I find mistakes made by a classifier?")
```
